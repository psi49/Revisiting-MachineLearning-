```python
feature = df.drop('Class', axis = 1)
target = df['Class']
```

```python
# Standardize the feature_space values 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
sc = StandardScaler()
feature = sc.fit_transform(feature)
# Reshape the target variable 
target=df['Class'].values.reshape(-1,1)
print(target.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size = 0.26, random_state = 0, stratify = target)
```

Train a Model using RandomForest(Classifier) and RandomSearch (For hyperparameter tunning)  credit card fraud detection

```python
from sklearn.metrics import classification_report, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier(random_state = 42)
from pprint import pprint
# Look at parameters used by our current Random Forest
print('Parameters of Random Forest in use:\n')
pprint(rf.get_params())
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 30, num = 3)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(20, 60, num = 4)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [3, 5]
# Minimum number of samples required at each leaf node

min_samples_leaf = [2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Method for selecting criterion
criterion = ['gini', 'entropy']

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'criterion': criterion,
               'bootstrap': bootstrap}
print('Parameters opted in the Random Grid to find best hyperparamerters fot the Random Forest: \n')
pprint(random_grid) 
```

```python
# First create the base model to tune
# Random search of parameters, using 3 fold cross validation, 
rf_randomsearch = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,
                               n_iter = 100, cv = 3, verbose=2)
# Fit the random search model
rf_randomsearch.fit(x_train, y_train)
```

```python
print("\n The best estimator across ALL searched params:\n", rf_randomsearch.best_estimator_) 
print("\n The best score across ALL searched params:\n",rf_randomsearch.best_score_)
print("\n The best parameters across ALL searched params:\n",rf_randomsearch.best_params_)
```

```python
rf_randomsearch_prediction = rf_randomsearch.predict(x_test)
_conf_matrix = confusion_matrix(y_test, rf_randomsearch_prediction)
print("Confusion matrix of Random Forest:\n", _conf_matrix)
print("Classification Report: \n {}".format (classification_report(y_test, rf_randomsearch_prediction)))
```
